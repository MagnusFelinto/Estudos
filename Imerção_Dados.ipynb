```python
g = dados[dados['tratamento'] == 'com_droga'].loc[:, 'g0':'g-49']
c = dados[dados['tratamento'] == 'com_droga'].loc[:,'c-0':'c-49']

x = pd.concat([g,c], axis=1)
```

```python
x
```

```python
corr = x.corr().round(2)
corr
```

```python
corr_filtrado = corr.loc[:'g-49','c-0':]
corr_filtrado.shape
```

```python
mask = np.triu(np.ones_like(corr_filtrado, dtype=bool))
#a variável mask é um triangulo superior (np.triu) de uma array das mesmas dimensões que nossa matrix de correlação
#e seus valores são do tipo booleano e iguais a 1 (True)
f, ax = plt.subplots(figsize=(11, 9)) #nesta etapa é selecionada a área que o gráfico irá ocupar
cmap = sns.diverging_palette(230, 20, as_cmap=True) #nesta etapa é selecionada a paleta de cores para a criação da diferença de cor, indo da por 230 até a 20
sns.heatmap(corr_filtrado, mask=mask, cmap=cmap, center = 0, linewidths=.5, cbar_kws={'shrink': .5}) #neste caso é construído o mapa de calor entre as relações.
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

x = dados_comb.select_dtypes('float64')
y = dados_comb['ativo_moa']
x_treino, x_teste, y_treino, y_test = train_test_split(x,y, test_size = 0.2, stratify=y, random_state=300)
modelo_rlogistica= LogisticRegression(max_iter=1000)
modelo_rlogistica.fit(x_treino,y_treino)
modelo_rlogistica.score(x_teste,y_test)
```

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score

modelo_dummy = DummyClassifier('most_frequent')
modelo_dummy.fit(x_treino, y_treino)

previsao_dummyt = modelo_dummy.predict(x_teste)
accuracy_score(y_test, previsao_dummyt)
```

```python
from sklearn.tree import DecisionTreeClassifier

x = dados_comb.select_dtypes('float64')
y = dados_comb['ativo_moa']
x_treino, x_teste, y_treino, y_test = train_test_split(x,y, test_size = 0.2, stratify=y, random_state=300)
teste=[]
treino=[]
for i in range(1,20):
  modelo_arvore = DecisionTreeClassifier(max_depth=i)
  modelo_arvore.fit(x_treino, y_treino)
  teste.append(modelo_arvore.score(x_teste,y_test))
  treino.append(modelo_arvore.score(x_treino,y_treino))
```
